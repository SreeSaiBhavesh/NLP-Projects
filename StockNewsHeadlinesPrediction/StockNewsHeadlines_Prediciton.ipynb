```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Data.csv',encoding = "ISO-8859-1")
```

We can see our dataset has 25 headlines for a day

```python
train_data = data[data['Date']<'20150101']
test_data = data[data['Date']>'20141231']
```

```python
train_data.shape, test_data.shape
```

```python
df = train_data.iloc[:,2:27]
```

```python
df
```

Cleaning the text data

```python
data.replace("[^a-zA-Z]"," ", inplace=True)
```

```python
data.head()
```

```python
new_index=[str(i) for i in range(25)]
```

```python
new_index
```

```python
df.columns = new_index
df
```

Converting data into lowercase

```python
for index in new_index:
    df[index] = df[index].str.lower()
df.head()
```

Performing Data Wrangling and Merging

```python
# Data of 2nd row
df.iloc[1,0:25]
```

```python
headlines = []
for i in df.iloc[1,0:25]:
    headlines.append(i)
' '.join(headlines)
```

```python
' '.join([str(i) for i in df.iloc[1,0:25]])
```

```python
headlines = []
for row in range(0,len(df)):
    headlines.append(' '.join([str(i) for i in df.iloc[row,0:25]]))
```

```python
headlines[2:5]
```

Applying Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
cv = CountVectorizer(ngram_range=(2,2))
traindata_x=cv.fit_transform(headlines)
```

Implementing RandomForestClassifer

```python
rf = RandomForestClassifier(n_estimators=200, criterion='entropy')
```

```python
rf.fit(traindata_x, train_data['Label'])
```

```python
test_transform = []
for row in range(0,len(test_data)):
    test_transform.append(' '.join([str(i) for i in test_data.iloc[row,2:27]]))
```

```python
testData = cv.transform(test_transform)
```

```python
predictions = rf.predict(testData)
predictions
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(test_data['Label'], predictions)
cm
```

```python
import matplotlib.pyplot as plt
plt.imshow(cm,cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
labels = ['positive', 'negative']
tick_marks = np.arange(len(labels))
plt.xticks(tick_marks,labels)
plt.yticks(tick_marks,labels)
plt.tight_layout()
plt.xlabel('True Labels')
plt.ylabel('Predicted Labels')
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
accuracy_score(test_data['Label'], predictions)
```

```python
report = classification_report(test_data['Label'], predictions)
```

```python
print(report)
```

```python
def plot_confusion_matrix(cm,title='Confusion Matrix'):
    import matplotlib.pyplot as plt
    plt.imshow(cm,cmap=plt.cm.Blues)
    plt.title(title)
    plt.colorbar()
    labels = ['positive', 'negative']
    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks,labels)
    plt.yticks(tick_marks,labels)
    plt.tight_layout()
    plt.xlabel('True Labels')
    plt.ylabel('Predicted Labels')
```

```python
plot_confusion_matrix(cm2)
```
